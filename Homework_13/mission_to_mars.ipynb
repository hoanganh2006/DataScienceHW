```python
from bs4 import BeautifulSoup as bs
import requests
import time
import pymongo
from splinter import Browser
import pandas as pd
```

```python
conn = 'mongodb://localhost:27017'
client = pymongo.MongoClient(conn)
executable_path = {'executable_path': '/usr/local/bin/chromedriver'}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
url = "https://mars.nasa.gov/news/?page=0&per_page=40&order=publish_date+desc%2Ccreated_at+desc&search=&category=19%2C165%2C184%2C204&blank_scope=Latest"
```

```python
response = requests.get(url)
soup = bs(response.text, 'html.parser')
```

```python
results = soup.find_all('div', class_="content_title")
for result in results: 
    news_title = result.a.text
    print(news_title)
```

```python
paragraph_results = soup.find_all('div', class_= 'rollover_description_inner')
for paragraph in paragraph_results: 
    news_p = paragraph.text
    print(news_p)
```

```python
image_url = 'https://www.jpl.nasa.gov/spaceimages/?search=&category=Mars'
image_response = requests.get(image_url)
image_soup = bs(image_response.text, 'html.parser')
```

```python
image_result = image_soup.find_all('a', class_= 'fancybox')
image_result
```

```python
for image in image_result:
    featured_image_url = image['data-fancybox-href']
    featured_image_url = 'https://www.jpl.nasa.gov/' + featured_image_url
    browser.visit(featured_image_url)
    print(featured_image_url)
```

```python
twitter_url = 'https://twitter.com/marswxreport?lang=en'
twitter_response = requests.get(twitter_url)
twitter_soup = bs(twitter_response.text, 'html.parser')
```

```python
mars_results = twitter_soup.find_all('p', class_= 'TweetTextSize TweetTextSize--normal js-tweet-text tweet-text')
for mars_weather in mars_results: 
    print(mars_weather.text)
```

```python
fact_url = 'https://space-facts.com/mars/'
fact_response = requests.get(fact_url)
fact_soup = bs(fact_response.text, 'html.parser')
```

```python
links = fact_soup.find('ul', class_= 'sub-menu')
links = links.find_all('a')
data = []
for link in links: 
    fact_link = link['href']
    fact = requests.get(fact_link)
    link_soup = bs(fact.content, 'lxml')
    table = link_soup.find_all('table')
    df = pd.read_html(str(table))   
    print(df[0])
```

```python
hemisphere_url = 'https://astrogeology.usgs.gov/search/results?q=hemisphere+enhanced&k1=target&v1=Mars'
hemisphere_response = requests.get(hemisphere_url)
hemisphere_soup = bs(hemisphere_response.text, 'html.parser')
print(hemisphere_soup.prettify())
```

```python
hemisphere_results = hemisphere_soup.find_all('a', class_ = 'itemLink product-item')
hemisphere_dict = {}
hemisphere_list = []
for hemisphere in hemisphere_results: 
    hemisphere_title = hemisphere.h3.text
    hemisphere_link = 'https://astrogeology.usgs.gov'+ hemisphere['href']
    full_img_response = requests.get(hemisphere_link)
    full_img_soup = bs(full_img_response.text, 'html.parser')
    full_img = full_img_soup.find('img', class_="wide-image")
    full_img_link = 'https://astrogeology.usgs.gov' + full_img['src']
    hemisphere_dict = {'title': hemisphere_title, 'img_url': full_img_link}
    hemisphere_list.append(hemisphere_dict)
hemisphere_list
```
